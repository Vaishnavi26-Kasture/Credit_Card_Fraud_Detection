## Model Prediction

```python
from sklearn.ensemble import IsolationForest
from sklearn.neighbors import LocalOutlierFactor
from sklearn.svm import OneClassSVM

# Step 1: Define necessary variables
X = card_data.drop('Class', axis=1)
outlier_fraction = card_data['Class'].value_counts()[1] / len(card_data)
state = 42

# Step 2: Define models
Classifier = {
    "Isolation Forest": IsolationForest(
        n_estimators=100,
        max_samples=len(X),
        contamination=outlier_fraction,
        random_state=state,
        verbose=0
    ),
    "Local Outlier Factor": LocalOutlierFactor(
        n_neighbors=20,
        algorithm='auto',
        leaf_size=30,
        metric='minkowski',
        p=2,
        contamination=outlier_fraction
    ),
    "Support Vector Machine": OneClassSVM(
        kernel='rbf',
        degree=3,
        gamma=0.1,
        nu=0.05,
        max_iter=-1
    )
}
```

```python
type(Classifier)
```

```python
from sklearn.metrics import classification_report, accuracy_score

# Define target labels (assuming 'Class' is the label)
Y = card_data['Class']
X = card_data.drop('Class', axis=1)

n_outliers = len(fraud)

for i, (clf_name, clf) in enumerate(Classifier.items()):
    print(f"\nRunning {clf_name}...")
    
    if clf_name == "Local Outlier Factor":
        y_pred = clf.fit_predict(X)
        scores_prediction = clf.negative_outlier_factor_
        
    elif clf_name == "Support Vector Machine":
        clf.fit(X)
        y_pred = clf.predict(X)
        
    else:  # Isolation Forest
        clf.fit(X)
        y_pred = clf.predict(X)
    
    # Convert predictions: 1 -> 0 (normal), -1 -> 1 (outlier)
    y_pred = [1 if x == -1 else 0 for x in y_pred]
    
    n_errors = (y_pred != Y).sum()
    
    print(f"{clf_name}: Number of errors = {n_errors}")
    print("Accuracy Score:", accuracy_score(Y, y_pred))
    print("Classification Report:")
    print(classification_report(Y, y_pred))
```
